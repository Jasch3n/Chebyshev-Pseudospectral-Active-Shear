```python
import sys
sys.path.append("../core-scripts")
import zeroShearSpectrum2D
import zeroShearSpectrum3D
from Utilities import prune_evs
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.lines as mlines

# Colors for plotting!
cs = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']
```

```python
def compute_zero_shear_specs(k,a,x=0.01,M_low=50,M_high=100):
    low_res_2D = np.array(zeroShearSpectrum2D.spectrum(k,a,_ell_over_W_squared=x,M=M_low), dtype=np.complex128)
    high_res_2D = np.array(zeroShearSpectrum2D.spectrum(k,a,_ell_over_W_squared=x,M=M_high), dtype=np.complex128)
    low_res_3D, _ = zeroShearSpectrum3D.spectrum(k,a,_ell_over_W_squared=x,M=M_low)
    high_res_3D, _ = zeroShearSpectrum3D.spectrum(k,a,_ell_over_W_squared=x,M=M_high)
    low_res_3D = np.array(low_res_3D, dtype=np.complex128)
    high_res_3D = np.array(high_res_3D, dtype=np.complex128)
    pruned_2D = prune_evs(low_res_2D, high_res_2D)
    pruned_3D = prune_evs(low_res_3D, high_res_3D)
    return (pruned_2D, pruned_3D)
```

```python
# Evaluate the spectrum at zero (very small) activity at different wavenumber values
kl,kr,kn = 0.001, 2, 8
ks = np.linspace(kl,kr,kn)
spec_k_2D = np.zeros(kn,dtype=np.complex128)
spec_k_3D = np.zeros(kn,dtype=np.complex128)
a = 10

for i in range(kn):
    print("{}/{} k values computed".format(i+1,kn))
    k = ks[i]
    pruned_2D, pruned_3D = compute_zero_shear_specs(k,a)
    spec_k_2D[i] = pruned_2D[0]
    spec_k_3D[i] = pruned_3D[0]
# np.savetxt("data/2D-3D-zero-shear-comp/k-spectrum/spec-k-2D.txt", spec_k_2D)
# np.savetxt("data/2D-3D-zero-shear-comp/k-spectrum/spec-k-3D.txt", spec_k_3D)
```

```python
def plot_stability(acts, ks, evs):
    an = len(acts)
    kn = len(ks)
    gr_th_zr = np.real(evs) > 0
    for i in range(an):
        for j in range(kn):
            c = "red" if gr_th_zr[i][j] else "blue"
            plt.plot(ks[j],acts[i],".",color=c, picker=True)
            plt.xlabel("$k$")
            plt.ylabel("$a$")
            blue_dot = mlines.Line2D([], [], color='blue', marker='.', linestyle='None',
                                  markersize=10, label='stable')
            red_dot = mlines.Line2D([], [], color='red', marker='.', linestyle='None',
                                  markersize=10, label='unstable')
            plt.legend(handles=[blue_dot, red_dot], loc='upper right')
```

```python
def find_crit_acts(ev_mat):
    # Finding the computational values of critical activity
    m,n = ev_mat.shape
    crit_a = np.zeros(n)

    for i in range(n):
        for j in range(m-1):
            ev = ev_mat[j][i]
            next_ev = ev_mat[j+1][i]
            if np.real(ev) * np.real(next_ev) < 0:
                print(ev, next_ev)
                crit_a[i] = acts[j]
                break
    return crit_a
```

# Effect of changing $l^2/W^2$ (increasing/decreasing correlation length and/or channel width)

```python
# Fix activity, evaluate the spectrum at different values of k and l^2/W^2
xs = np.linspace(0.01, 0.2, 7)
ks = [0.3, 0.6, 1, 2, 3, 7, 16]
a = 1
spec_ell_2D_k = np.zeros((len(ks), len(xs)), dtype=np.complex128)
spec_ell_3D_k = np.zeros((len(ks), len(xs)), dtype=np.complex128)
for i in range(len(ks)):
    print(f"{i+1}/{len(ks)}")
    for j in range(len(xs)):
        pruned_2D, pruned_3D = compute_zero_shear_specs(ks[i],a,x=xs[j])
        spec_ell_2D_k[i][j] = pruned_2D[0]
        spec_ell_3D_k[i][j] = pruned_3D[0]
```

```python
np.savetxt("data/2D-3D-zero-shear-comp/l-over-W-sq-spectrum/spec-ell-2D-k-high-res.txt", spec_ell_2D_k)
np.savetxt("data/2D-3D-zero-shear-comp/l-over-W-sq-spectrum/spec-ell-3D-k-high-res.txt", spec_ell_3D_k)
```

```python
# Here we find that the 3D system stabilizes SLOWER than the 2D system
# as the correlation length is increased. But at large wavenumbers, the difference doesn't seem
# to be important
xs = np.linspace(0.01, 0.2, 7) # these are values of l^2/W^2
ks = [0.3, 0.6, 1, 2, 3, 7, 16]

fig, ax = plt.subplots()
lines_2D = []
lines_3D = []

spec_ell_2D_k = np.loadtxt("data/2D-3D-zero-shear-comp/l-over-W-sq-spectrum/spec-ell-2D-k-high-res.txt",dtype=np.complex128)
spec_ell_3D_k = np.loadtxt("data/2D-3D-zero-shear-comp/l-over-W-sq-spectrum/spec-ell-3D-k-high-res.txt",dtype=np.complex128)


for k_indx in [0,1,2]:
    spec2D_k = np.real(spec_ell_2D_k[k_indx])
    spec3D_k = np.real(spec_ell_3D_k[k_indx])
    line_2D, = plt.plot(xs, spec2D_k,'--', label="$k={}$".format(ks[k_indx]), color=cs[k_indx])
    line_3D, = plt.plot(xs, spec3D_k, label="$k={}$".format(ks[k_indx]), color=cs[k_indx])
    plt.xlabel("$l^2/W^2$")
    plt.ylabel("Re($\sigma$)")
    lines_2D.append(line_2D)
    lines_3D.append(line_3D)
plt.title("Normalized Re($\sigma$) vs. $l^2/W^2$ at different values of $k$, $a$=1")
legend_2D = ax.legend(handles=lines_2D, loc=(1.04,0.55), title="2D")
ax.add_artist(legend_2D)
ax.legend(handles=lines_3D, loc=(1.04,0), title="3D")
plt.title("Re($\sigma$) vs. $l^2/W^2$ at different values of $k$, $a=1$")
```

```python
ks = np.load("data/2D-3D-zero-shear-comp-fine/ks.npy")
acts = np.load("data/2D-3D-zero-shear-comp-fine/acts.npy")
```

```python
# First we check that the pruning algorithm doesn't affect the correctness of the 2D calculations
spec_2D_low = []
spec_2D_clean = []
for i in range(len(ks)):
    for j in range(len(acts)):
        spec_data = np.load(f'data/2D-3D-zero-shear-comp-fine/k{i}-a{j}.npy')
        spec_2D_low.append(spec_data[0,0])
        spec_2D_clean.append(spec_data[1,0])
fig,ax = plt.subplots(1,2)
ax[0].plot(np.arange(len(spec_2D_low)), np.real(spec_2D_low) - np.real(spec_2D_clean))
ax[1].plot(np.arange(len(spec_2D_low)), np.imag(spec_2D_low) - np.imag(spec_2D_clean))
ax[0].set_ylim([-0.3,1])
ax[1].set_ylim([-0.3,1])
ax[0].set_title("Difference Between Real Parts")
ax[1].set_title("Difference Between Imaginary Parts")
```

```python
fig, ax = plt.subplots(1,2)
for i in range(len(ks)):
    for j in range(len(acts)):
        spec_data = np.load(f'data/2D-3D-zero-shear-comp-fine/k{i}-a{j}.npy')
        ax[0].plot(ks[i], acts[j], marker='o', c='red' if np.real(spec_data[1,0])>0 else 'blue')
        ax[1].plot(ks[i], acts[j], marker='o', c='red' if np.real(spec_data[2,0])>0 else 'blue')
```

```python
crit_acts_2D = []
crit_acts_3D = []
for i in range(len(ks)):
    for j in range(len(acts) - 1):
        spec_data = np.load(f'data/2D-3D-zero-shear-comp-fine/k{i}-a{j}.npy')
        spec_data_next = np.load(f'data/2D-3D-zero-shear-comp-fine/k{i}-a{j+1}.npy')
        if (np.real(spec_data_next[1,0])>0 and np.real(spec_data[1,0])<0):
            crit_acts_2D.append(acts[j])
        if (np.real(spec_data_next[2,0])>0 and np.real(spec_data[2,0])<0):
            crit_acts_3D.append(acts[j])
plt.plot(ks, crit_acts_2D, label='2D')
plt.plot(ks, crit_acts_3D, label='3D')
plt.legend()
plt.title("2D and 3D Critical Activities at Zero Shear")
plt.xlabel("$k$")
plt.ylabel("$a$")
```
