```python
import sys
sys.path.append("../core-scripts")
from zeroShearSpectrum3D import spectrum
import numpy as np
import matplotlib.pyplot as plt
from cheb import cheb
```

```python
k,a = 1.25, 1.67
evals_low, evecs_low = spectrum(k,a,M=50)
evals_high, evecs_high = spectrum(k,a,M=100)
np.savetxt("data/manual-implementation/evals_M_50.txt", evals_low)
np.savetxt("data/manual-implementation/evals_M_100.txt", evals_high)
np.savetxt("data/manual-implementation/evecs_M_50.txt", evecs_low)
np.savetxt("data/manual-implementation/evecs_M_100.txt", evecs_high)
```

```python
M = 50
M_low, M_high = M, 2*M
D1_low, ygl_low = cheb(M_low)
D1_high, ygl_high = cheb(M_high)

Qxx_low = plt.figure()
mode_indices = [0, 1, 4, 10, 21, 41, 90]
for i in mode_indices:
    Qxx_re = np.real(evecs_low[i, 3*M_low:4*M_low])
#     Qxx_im = np.imag(evecs_low[i, 3*M:4*M])
    plt.plot(ygl_low, Qxx_re, marker='x',linestyle='--')
#     plt.plot(ygl_low, Qxx_im)

Qxx_high = plt.figure()
for i in mode_indices:
    Qxx_re = np.real(evecs_high[i, 3*M_high:4*M_high])
    plt.plot(ygl_high, Qxx_re, marker='x',linestyle='--')
```

```python
Vz_low = plt.figure()
mode_indices = [10]
for i in mode_indices:
    Vz_re = np.real(evecs_low[i, 2*M_low:3*M_low])
    Vz_im = np.imag(evecs_low[i, 2*M_low:3*M_low])
    plt.plot(ygl_low, Vz_re, marker='x',linestyle='--')
    plt.plot(ygl_low, Vz_im, marker='x',linestyle='--')

Vz_high = plt.figure()
for i in mode_indices:
    Vz_re = np.real(evecs_high[i, 2*M_high:3*M_high])
    Vz_im = np.imag(evecs_high[i, 2*M_high:3*M_high])
#     plt.plot(ygl_high, Vz_re, marker='x',linestyle='--')
#     plt.plot(ygl_high, Vz_im, marker='x',linestyle='--')
```
