```python
"""
Evaluate spectra on a grid of gammadot*tau and k
"""

from shear2D import ev, max_ev_kg_grid
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.lines as mlines


def graph_spectra_gamma_k_grid(k_specs, gd_specs, tau, tau_a):
    kl,kr,k_step = k_specs[0], k_specs[1], k_specs[2]
    gdl,gdr,gd_step = gd_specs[0],gd_specs[1],gd_specs[2]
    ks = np.arange(kl, kr, k_step)
    gds = np.arange(gdl, gdr, gd_step)
    kv, gv = np.meshgrid(ks, gds)

    # denote the matrix sizes as NxM
    Nk,Mk = len(kv),len(kv[0])
    Ng,Mg = len(gv),len(gv[0])

    mat = max_ev_kg_grid(ks,gds,tau,tau_a)
    for i in range(Nk):
        for j in range(Mk):
            k = kv[i][j]
            c = "red" if mat[i][j] > 0 else "blue"
            plt.plot(kv[i][j],gv[i][j],".",color=c, picker=True)
    plt.xlabel("$k$")
    plt.ylabel("$\dot\gamma$")
    blue_dot = mlines.Line2D([], [], color='blue', marker='.', linestyle='None',
                          markersize=10, label='stable')
    red_dot = mlines.Line2D([], [], color='red', marker='.', linestyle='None',
                          markersize=10, label='unstable')
    plt.legend(handles=[blue_dot, red_dot])
    plt.title("Stability Plot in the $\dot\gamma$-$k$ plane, $\\tau_a$={}".format(tau_a))
    plt.show()
```

```python
"""THIS CODE BLOCK IS USED FOR TESTING"""
# parameters
tau, tau_a = 1, 0.1
gdl, gdr, gd_step = 1, 2, 0.5
kl, kr, k_step = 1, 2, 0.5
   
# make stability plot
graph_spectra_gamma_k_grid([kl,kr,k_step],[gdl,gdr,gd_step],tau,tau_a)
```
